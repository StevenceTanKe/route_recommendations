# 景点的推荐
第四步我们得到了八个标签，每个标签下面有若干个景点。接下来需要解决的是同一个标签下各景点的推荐顺序问题以及多标签下景点的推荐顺序问题。
- 对于同一标签，我选择*TF-IDF值*作为某标签下景点的排名依据，因为它的大小能够体现它在文本集合中的某一个文档里的重要性。
- 而对于多标签，因为用户是基于自己的兴趣选择标签，选择的顺序不同，其重要性也不同，因此，考虑让用户自定义所选择各标签的权重。

## 单标签下的景点排名

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# files = [
#     '../data/processed/attractions_topic_clustering/Square.xlsx',
#     '../data/processed/attractions_topic_clustering/Park.xlsx',
#     '../data/processed/attractions_topic_clustering/Street.xlsx',
#     '../data/processed/attractions_topic_clustering/Museum.xlsx',
#     '../data/processed/attractions_topic_clustering/Performance.xlsx',
#     '../data/processed/attractions_topic_clustering/Western-Church.xlsx'
#     '../data/processed/attractions_topic_clustering/Chinese.xlsx',
#     '../data/processed/attractions_topic_clustering/Landmark.xlsx',
# ]
comment_processed_file_path = '../data/processed/attractions_reviews.xlsx'

# 使用pandas直接读取并合并所有文件
df_comment = pd.read_excel(comment_processed_file_path)

# 使用TfidfVectorizer计算TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df_comment['comment'])
feature_names = tfidf_vectorizer.get_feature_names_out()

# 主题词
topics = {
    'Square': ['广场', '喷泉', '活动'],
    'Park': ['公园', '花园', '园林', '自然'],
    'Street': ['街', '街道', '街区', '街市', '老街', '新街'],
    'Museum': ['博物馆', '陈列馆', '展厅', '场馆', '展馆', '图书馆'],
    'Performance': ['表演', '演出', '电影', '观光', '烟花', '现场', '小朋友', '互动'],
    'Western-Church': ['西式', '教堂', '葡萄牙', '圣堂', '罗马', '古罗马', '灯塔', '修院', '牌坊', '浪漫'],
    'Chinese': ['中式', '庙', '庙宇', '寺庙', '佛教', '观音', '岭南', '本地人', '园林', '新华'],
    'Landmark': ['地标', '打卡', '标志', '标志性']
}

# 创建一个字典来保存关键词索引
keyword_index = {word: i for i, word in enumerate(feature_names) if word in sum(topics.values(), [])}

# 计算每个景点每个标签的最高TF-IDF值
for label, keywords in topics.items():
    # 只处理存在于特征名称中的关键词
    valid_keywords = [k for k in keywords if k in keyword_index]
    # 获取这些关键词的索引
    indices = [keyword_index[k] for k in valid_keywords]
    # 计算每条评论中这些关键词的最高TF-IDF值
    max_tfidf = tfidf_matrix[:, indices].max(axis=1).toarray().flatten()
    df_comment[label] = max_tfidf

# 对每个标签下的景点进行排序
sorted_attractions = {label: df_comment.groupby('Attraction').agg({label: 'max'}).sort_values(by=label, ascending=False)
                      for label in topics}

# 显示排序结果
print("Sorting results of attractions under each topic:")
for label, df in sorted_attractions.items():
    print(f'\n Label: {label}')
    print(df)
```

在`attractions_information.xlsx`文件中新增八列，列名为八个主题，表格中的值为各景点在各主题下的*最高TF-IDF值*。

```python
attraction_file_path_manually = '../data/processed/manually_processed_attractions_information/attractions_information_before_tfidf.xlsx'
attraction_file_path = '../data/processed/attractions_information.xlsx'
df_attraction = pd.read_excel(attraction_file_path_manually)

df_tfidf  = pd.DataFrame()

# 遍历每个标签下的DataFrame，将其合并到df_tfidf中
for label, df in sorted_attractions.items():
    # 如果df_tfidf为空，则直接添加第一个DataFrame
    if df_tfidf.empty:
        df_tfidf = df.rename(columns={label: label})
    else:
        # 否则，将新的DataFrame与final_df进行合并
        df_tfidf = df_tfidf.join(df.rename(columns={label: label}), how='left')

# 补充可能因为缺少数据而出现的NaN值，用0替代
df_tfidf.fillna(0, inplace=True)

df_attraction = pd.merge(df_attraction, df_tfidf, on='Attraction', how='outer')

df_attraction.to_excel(attraction_file_path, index=False)
```

## 多标签下的景点排名
让用户自定义所选择第i个标签的权重$w_i$ $i=1,2,\cdots,8$, 其中未被选中的标签其权重为0；第i个标签下第j个景点的TF-IDF值记为$t_{ij},\quad j=1,2,\cdots,49$, 若第j个景点没有出现在第i个标签下，记其 $t_{ij}=0$,

则经过加权计算，在n个标签下各景点的得分为$S_k=\sum_{i=1}^8w_i\cdot t_{ik},\quad k=1,2,\cdots,49$。

实例展示：用户选择的标签为*西式-教堂*、*地标*，其权重分别为*0.2*和*0.8*。根据上述公式计算可得打分排名结果如下：

```python
# 用户定义的标签及其权重
label_weights = {
    'Western-Church': 0.2,
    'Landmark': 0.8
}

# 执行加权得分计算和排序
df_attraction['Weighted_Score'] = 0
for label, weight in label_weights.items():
    df_attraction['Weighted_Score'] += df_attraction.get(label, 0) * weight
df_recommendation = df_attraction.sort_values(by='Weighted_Score', ascending=False).head(20)

# 输出结果
print("Ranking of attractions under user-selected tags:")
print(df_recommendation[['Attraction', 'Weighted_Score']].to_string(index=False))
```

# 旅游路线的推荐
在选出最适合用户的景点群后，考虑为用户设计最优的游览路线，该部分通过设定两种不同的场景进行展示。

## 多日游总体路线规划
**情境设定** \
因为澳门很多景点都是以观光为主，不会停留太久，所以我们选择二十个景点来规划路线。不考虑机场、火车站、酒店、餐厅的位置，不考虑旅行天数限制，仅考虑游完二十个景点的最短路径路线，游客选择的标签为*西式-教堂*、*地标*。主要为了展示多个标签下的景点选择问题以及景点确定后的路线规划。

### 获取推荐景点的经纬度数据并计算距离矩阵
得到给用户推荐的景点，并使用经纬度数据来计算两点间的球面距离。\
\
因为之前已经获取过景点的地址，所以这里直接调用`attraction.xlsx`文件获取景点的经纬度。设第i个景点与第j个景点的经纬度坐标分别为 $(x_i,y_i)$ 和 $(x_j,y_j)$, 则这两点的距离为：

$$
d_{ij}=R\cdot\arccos\left[\cos\left(x_1-x_2\right)\cdot\cos y_1\cdot\cos y_2+\sin y_1\cdot\sin y_2\right]
$$

其中R为地球半径，从而得到距离矩阵D，$(d_{ij})_{10\times10}$。


```python
import numpy as np

# 加载经纬度数据
df_coords = df_attraction.loc[df_recommendation.index]  # 只获取推荐景点的坐标

# 计算地球上两点间的距离
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # 地球半径（公里）
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

# 创建距离矩阵
distance_matrix = np.zeros((len(df_coords), len(df_coords)))
for i, (idx1, row1) in enumerate(df_coords.iterrows()):
    for j, (idx2, row2) in enumerate(df_coords.iterrows()):
        distance_matrix[i, j] = haversine(row1['lat'], row1['lng'], row2['lat'], row2['lng'])
```

### 蚁群算法得出TSP问题最优路线规划
*蚁群算法 (ACA)* 是对自然界蚂蚁的寻径方式进行模拟而得出的一种仿生算法，由大量蚂蚁组成的蚁群集体行为便表现出一种信息正反馈现象：某一路径上走过的蚂蚁越多，则后来者选择该路径的概率就越大。本质上是进化算法中的一种启发式全局优化算法。

```python
import pants

# 创建距离矩阵的计算函数
def distance(i, j):
    return distance_matrix[i][j]

# 创建世界和蚂蚁
world = pants.World(range(len(distance_matrix)), distance)
# 定义蚁群算法的参数
solver = pants.Solver(
    alpha=1.0,  # 信息素重要性
    beta=4.0,   # 启发式信息的重要性，通常设置更大的beta以偏好短距离
    rho=0.5,    # 信息素的蒸发率
    ant_count=100,  # 蚂蚁的数量
    limit=1000,  # 迭代次数
)

# 求解TSP问题
solution = solver.solve(world)

# 输出最优路径和总距离
optimal_path = solution.tour
optimal_distance = solution.distance
print("Optimal path:", optimal_path + [optimal_path[0]])
print("Optimal distance:", optimal_distance)
```

经过计算，这个*哈密顿环*的最短路径大致为：*大三巴街 -> 恋爱巷 -> 大三巴牌坊 -> 天主教圣安多尼堂 -> 疯堂斜巷 -> 塔石广场 -> 东望洋山 -> 东望洋炮台及灯塔 -> 金莲花广场 -> 澳门渔人码头 -> 澳门塔 -> 西望洋山 -> 主教山小堂 -> 圣老楞佐堂 -> 圣若瑟修院藏珍馆 -> 岗顶前地 -> 议事亭前地 -> 邮政总局 -> 圣母诞辰主教座堂 -> 玫瑰圣母堂 -> 大三巴街*  (起点可以为任意一个景点)，其最短距离为*7.1098公里*。

```python
import matplotlib.pyplot as plt

# 获取最优路径的景点坐标
optimized_coords = [(df_coords.iloc[idx]['lng'], df_coords.iloc[idx]['lat']) for idx in optimal_path]

# 解压坐标列表
lngs, lats = zip(*optimized_coords)

# 绘制路径图
plt.figure(figsize=(10, 6))
plt.plot(list(lngs) + [lngs[0]], list(lats) + [lats[0]], 'o-', markersize=8, linewidth=2, label='Travel Path')
plt.title('Optimal Travel Route')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.legend()
plt.show()
```

### 优化路线
我们将找到最长的边并从路线中移除，以优化总体旅行路径。

由距离比较得，*澳门塔*与*澳门渔人码头*之间的距离最长，故可以考虑以*澳门塔*为起点、*澳门渔人码头*为终点或*澳门渔人码头*为起点、*澳门塔*为终点，其路线长度为*4.619612801770785公里*。

```python
# 找到最长的边
longest_dist = 0
longest_index = 0
for i in range(len(optimal_path) - 1):
    dist = distance_matrix[optimal_path[i], optimal_path[i + 1]]
    if dist > longest_dist:
        longest_dist = dist
        longest_index = i

# 移除最长边并显示新的旅行顺序
optimized_path = optimal_path[longest_index + 1:] + optimal_path[:longest_index + 1]
print("Optimized visit order:", optimized_path)
print("Optimized visit distance:", optimal_distance - longest_dist)
```

```python
# 从df_coords获取景点名称
attraction_names = df_coords['Attraction'].tolist()

# 生成优化后的景点访问顺序的名称
optimized_route_names = [attraction_names[idx] for idx in optimized_path]

# 打印最终路线
print("Final optimized route:")
print(" -> ".join(optimized_route_names))
```

## 一日游总体路线规划
**情境设定** \
早晨十点开始旅行，晚上十点结束，在景点沿线午餐一个小时、晚餐一个小时，不涉及起终点，用户在每个景点的用时为建议游览时间，景点间的交通方式为步行。

因为澳门半岛的景点比较集中，景点之间的路程不算很远，所以将步行作为交通工具。通过ChatGPT-4得到相应的速度——*步行： 5 km/h*。


**问题的分析** \
根据情境，约束条件为逛景点及景点间交通的时间为十个小时，每个景点都需要满足建议游览时间的限制。为了给游客提供良好的游览体验，应当让游客的交通时间尽量缩短，应当让游客游览尽可能多的景点。当用户能够走尽量多的景点时，由于总时间固定，意味着游览时间尽可能多，使得交通时间尽可能短，故可以将多个优化目标确定为*第一目标：景点数量尽可能地多*，*第二目标：景点评分尽量高*。

### 确定推荐景点
建议游览时间数据来源于携程网海外版，由于优化目标可知，选择时间区间的下限能够使得景点数目尽量多。然后通过之前的步骤确定要推荐的景点。

```python
df_attraction = pd.read_excel(attraction_file_path)
df_duration = pd.read_excel('../data/raw/attractions_information/attractions_recommendation_duration.xlsx')

# 获得计算用的游览时间
df_duration['tour_duration'] = df_duration['建议逗留时间'].apply(lambda x: float(x.split('-')[0].split('小')[0]))
df_attraction = pd.merge(df_attraction, df_duration[['Attraction', 'tour_duration']], on='Attraction', how='outer')
df_attraction.to_excel(attraction_file_path, index=False)
speed = 5

# 用户定义的标签及其权重
label_weights = {
    'Western-Church': 0.2,
    'Landmark': 0.8
}

# 执行加权得分计算和排序
df_attraction['Weighted_Score'] = 0
for label, weight in label_weights.items():
    df_attraction['Weighted_Score'] += df_attraction.get(label, 0) * weight
df_recommendation = df_attraction.sort_values(by='Weighted_Score', ascending=False).head(10)

print(df_recommendation[['Attraction', 'Weighted_Score']].to_string(index=False))
print(f"\n Stay duration: {df_recommendation['tour_duration'].sum()} km/h")
```

### 计算最短路径
先不考虑十个小时的约束条件，对这10个景点运用*蚁群算法*计算该哈密顿环的最短路径。然后找到最长的边并从路线中移除，以优化总体旅行路径。

```python
# 加载经纬度数据
df_coords = df_attraction.loc[df_recommendation.index]

# 创建距离矩阵
distance_matrix = np.zeros((len(df_coords), len(df_coords)))
for i, (idx1, row1) in enumerate(df_coords.iterrows()):
    for j, (idx2, row2) in enumerate(df_coords.iterrows()):
        distance_matrix[i, j] = haversine(row1['lat'], row1['lng'], row2['lat'], row2['lng'])
        

world = pants.World(range(len(distance_matrix)), distance)
solver = pants.Solver(
    alpha=1.0,  # 信息素重要性
    beta=4.0,   # 启发式信息的重要性，通常设置更大的beta以偏好短距离
    rho=0.5,    # 信息素的蒸发率
    ant_count=100,  # 蚂蚁的数量
    limit=1000,  # 迭代次数
)

# 求解TSP问题
solution = solver.solve(world)

# 输出最优路径和总距离
optimal_path = solution.tour
optimal_distance = solution.distance

# 找到最长的边
longest_dist = 0
longest_index = 0
for i in range(len(optimal_path) - 1):
    dist = distance_matrix[optimal_path[i], optimal_path[i + 1]]
    if dist > longest_dist:
        longest_dist = dist
        longest_index = i

# 移除最长边并显示新的旅行顺序
optimized_path = optimal_path[longest_index + 1:] + optimal_path[:longest_index + 1]
optimized_distance = optimal_distance - longest_dist
print("Optimized visit order:", optimized_path)
print("Optimized visit distance:", optimized_distance)

total_tour_duration = optimized_distance / speed + df_recommendation['tour_duration'].sum()
print("Optimized tour duration:", total_tour_duration)
```

```python
# 获取最优路径的景点坐标
optimized_coords = [(df_coords.iloc[idx]['lng'], df_coords.iloc[idx]['lat']) for idx in optimized_path]

# 解压坐标列表
lngs, lats = zip(*optimized_coords)

# 绘制路径图
plt.figure(figsize=(10, 6))
plt.plot(list(lngs) + [lngs[0]], list(lats) + [lats[0]], 'o-', markersize=8, linewidth=2, label='Travel Path')
plt.title('Optimal Travel Route')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.legend()
plt.show()
```

### 时间约束
考虑十个小时的约束，采用遍历法，从排名靠后的景点开始，先尝试剔除一个景点，判断是否满足约束；若都不满足约束，则考虑剔除两个景点，以此类推。

```python
# 如果超出10小时，则尝试去掉某些景点
while total_tour_duration > 10:
    # 从排名低的景点开始移除
    df_recommendation = df_recommendation[:-1]
    # 重新计算路径和总时间
    df_coords = df_attraction.loc[df_recommendation.index]

    # 创建距离矩阵
    distance_matrix = np.zeros((len(df_coords), len(df_coords)))
    for i, (idx1, row1) in enumerate(df_coords.iterrows()):
        for j, (idx2, row2) in enumerate(df_coords.iterrows()):
            distance_matrix[i, j] = haversine(row1['lat'], row1['lng'], row2['lat'], row2['lng'])

    world = pants.World(range(len(distance_matrix)), distance)
    solver = pants.Solver(
        alpha=1.0,  # 信息素重要性
        beta=4.0,   # 启发式信息的重要性，通常设置更大的beta以偏好短距离
        rho=0.5,    # 信息素的蒸发率
        ant_count=100,  # 蚂蚁的数量
        limit=1000,  # 迭代次数
    )

    solution = solver.solve(world)

    optimal_path = solution.tour
    optimal_distance = solution.distance

    longest_dist = 0
    longest_index = 0
    for i in range(len(optimal_path) - 1):
        dist = distance_matrix[optimal_path[i], optimal_path[i + 1]]
        if dist > longest_dist:
            longest_dist = dist
            longest_index = i

    optimized_path = optimal_path[longest_index + 1:] + optimal_path[:longest_index + 1]
    optimized_distance = optimal_distance - longest_dist
    print("Optimized visit order:", optimized_path)
    print("Optimized visit distance:", optimized_distance)

    total_tour_duration = optimized_distance / speed + df_recommendation['tour_duration'].sum()
    print("Optimized tour duration:", total_tour_duration)

attraction_names = df_coords['Attraction'].tolist()
optimized_route_names = [attraction_names[idx] for idx in optimized_path]

print("Final optimized route:")
print(" -> ".join(optimized_route_names))
```
